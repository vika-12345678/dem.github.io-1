```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor
from pptx.enum.text import MSO_ANCHOR, MSO_AUTO_SIZE
from copy import deepcopy

# Создаем новую презентацию
prs = Presentation(r'/home/user/Загрузки/mkramn/Шаблон презентации МК РАМН.pptx')

# Определяем фирменные цвета из брендбука
COLOR_DARK_BLUE = RGBColor(0x03, 0x41, 0x85)  # #034185
COLOR_WHITE = RGBColor(0xFF, 0xFF, 0xFF)      # #FFFFFF
COLOR_BLACK = RGBColor(0x00, 0x00, 0x00)      # #000000
COLOR_GRAY = RGBColor(0xA8, 0xA8, 0xA7)       # #A8A8A7

# Удаляем слайд по индексу(номеру с нуля)
def delete_slide_by_index(presentation, index):
    """Удаляет слайд по индексу (0-based)."""
    xml_slides = presentation.slides._sldIdLst
    slides = presentation.slides

    if index >= len(slides):
        raise IndexError("Индекс слайда вне диапазона")

    # Получаем элемент XML для слайда по индексу
    slide_id = xml_slides[index]

    # Удаляем его из списка
    xml_slides.remove(slide_id)

def Cm(cm):
    return Inches(cm / 2.54)

# Функция для заполнения слайда
def create_content_slide(title_text, content_text):
    """Создает слайд с заголовком и содержимым."""
    slide = prs.slides.add_slide(prs.slides[2].slide_layout) # Копируем слайд

    # Удаляем все shape'ы
    for shape in list(slide.shapes):  # list() чтобы сделать копию
        sp = shape._element
        sp.getparent().remove(sp)
       
    # Добавляем заголовок
    title_box = slide.shapes.add_textbox(Inches(1), Inches(1.5), Inches(11), Inches(1))
    title_frame = title_box.text_frame
    title_frame.clear()
    p = title_frame.paragraphs[0]
    p.text = title_text
    p.font.name = "Montserrat"
    p.font.size = Pt(32)
    p.font.bold = True
    p.font.color.rgb = COLOR_DARK_BLUE
    p.alignment = PP_ALIGN.LEFT

    # Добавляем основной текст
    content_box = slide.shapes.add_textbox(Inches(1), Inches(2.5), Inches(11), Inches(3))
    content_frame = content_box.text_frame
    content_frame.clear()
    content_frame.word_wrap = True
    p = content_frame.paragraphs[0]
    p.text = content_text
    p.font.name = "Montserrat"
    p.font.size = Pt(20)
    p.font.color.rgb = COLOR_DARK_BLUE
    p.alignment = PP_ALIGN.LEFT    
    return slide




# --- СОЗДАНИЕ ПРЕЗЕНТАЦИИ ---

# Слайд 1: Титульный (первый слайд из шаблона)
slide = prs.slides[0]

# Добавляем центральный текст 
center_title = slide.shapes.add_textbox(Inches(2.2), Inches(2.5), Inches(11), Inches(2))
title_frame = center_title.text_frame
title_frame.clear()
p = title_frame.paragraphs[0]
p.text = "Миссия медсестры"
p.font.name = "Montserrat"
p.font.size = Pt(44)
p.font.bold = True
p.font.color.rgb = COLOR_DARK_BLUE
p.alignment = PP_ALIGN.RIGHT

center_subtitle = slide.shapes.add_textbox(Inches(2.2), Inches(4.5), Inches(11), Inches(1))
subtitle_frame = center_subtitle.text_frame
subtitle_frame.clear()
p = subtitle_frame.paragraphs[0]
p.text = "101С, Ромашкина Светлана"
p.font.name = "Montserrat"
p.font.size = Pt(32)
p.font.bold = True
p.font.color.rgb = COLOR_DARK_BLUE
p.alignment = PP_ALIGN.RIGHT

# Слайд 2: Что такое миссия медсестры?
create_content_slide(    
    "Что такое миссия медсестры?",
    "Миссия медсестры — это ее жизненное призвание, основная цель и смысл профессиональной деятельности. Это не просто работа, а служение, основанное на сострадании, ответственности и стремлении облегчить страдания."
)

# Слайд 3: Основные принципы миссии
create_content_slide(   
    "Основные принципы миссии",
    "• Забота и сострадание: Бескорыстная помощь пациенту, уважение к его достоинству.\n"
    "• Профессионализм: Постоянное стремление к совершенствованию знаний и навыков.\n"
    "• Ответственность: Осознание важности своей роли в жизни пациента и общества.\n"
    "• Защита: Быть адвокатом пациента, защищать его права и интересы."
)

# Слайд 4: Миссия и ценности колледжа
create_content_slide(    
    "Миссия и ценности колледжа",
    "Наш колледж готовит специалистов, чья миссия основана на наших корпоративных ценностях:\n"
    "• Уважение к пациенту и коллегам.\n"
    "• Развитие личных и профессиональных качеств.\n"
    "• Ответственность за результат своей работы.\n"
    "• Устремленность в будущее для внедрения лучших практик."
)

# Слайд 5: Заключение
create_content_slide(   
    "Заключение",
    "Миссия медсестры — это фундамент, на котором строится вся ее профессиональная жизнь. Это ежедневный выбор помогать, заботиться и нести свет в жизнь тех, кто в этом нуждается. Колледж гордится тем, что готовит таких специалистов."
)

# нужно конечный слайд переместить из 1 позиции в последнюю
slide = prs.slides.add_slide(prs.slides[1].slide_layout) # Копируем слайд
for shape in prs.slides[1].shapes:
        new_elm = deepcopy(shape.element)
        slide.shapes._spTree.insert_element_before(new_elm, 'p:extLst')

# нужно удалить слайды
indices_to_delete = [2, 1]  # в обратном порядке!
for idx in sorted(indices_to_delete, reverse=True):
    delete_slide_by_index(prs, idx)

# Сохраняем презентацию
prs.save(r'/home/user/Загрузки/2025/Сгенерированная презентация.pptx')
print("Презентация готова")
```
